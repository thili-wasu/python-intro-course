```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

class PontdeVilomara:
    def __init__(self, filename):
        self.df = pd.read_csv(filename, delimiter='\t', header=2)
        self.filename = filename

        # Extract latitude and longitude
        with open(filename, 'r') as file:
            first_line = file.readline().strip()
        data = first_line.split(' ')

        lat_lon = data[6]
        data2 = lat_lon.split(',')
        lat = data2[0]
        lon = data2[1]

        self.lat = lat
        self.lon = lon

        # Extract date
        with open(filename, 'r') as file:
            file.readline()  # Skip the first line
            second_line = file.readline().strip()
        data = second_line.split(' ')
        date = data[1]
        date_string = str(date)
        date_object = datetime.strptime(date_string, '%Y%m%d')
        self.date_only = date_object.date()

        # Convert time range to datetime
        def convert_time_range_to_datetimes(time_range_str, date_str=self.date_only):
            start_time_str, end_time_str = time_range_str.split(' - ')
            start_datetime = datetime.strptime(f'{date_str} {start_time_str}', '%Y-%m-%d %H:%M')
            end_datetime = datetime.strptime(f'{date_str} {end_time_str}', '%Y-%m-%d %H:%M')
            return start_datetime  # Return only start_datetime for now

        # Apply the function to the DataFrame
        if 'Periode[UTM]' in self.df.columns:
            self.df[['date']] = self.df['Periode[UTM]'].apply(lambda x: pd.Series(convert_time_range_to_datetimes(x)))
            self.df.drop(columns=['Periode[UTM]'], inplace=True)
        else:
            print(f"The 'Periode[UTM]' column does not exist in DataFrame from {filename}.")

# List of filenames
myfiles = [
    '/data/IMFSE/PythonCourse/PontdeVilomara/WeatherStations/wx_20220716_CastellouDeBages.csv',
    '/data/IMFSE/PythonCourse/PontdeVilomara/WeatherStations/wx_20220717_CastellouDeBages.csv'
]

# List to store DataFrames
df_all = []

# Process each file
for file in myfiles:
    # Create an instance of PontdeVilomara for each file
    pdVW = PontdeVilomara(file)
    # Append the processed DataFrame to the list
    df_all.append(pdVW.df)

# Concatenate all DataFrames
df_combined = pd.concat(df_all, ignore_index=True)

# Display the combined DataFrame
print("Combined DataFrame:")
print(df_combined)
```
