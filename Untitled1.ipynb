```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

class PontdeVilomara:
    def __init__(self, filename):
        # Load the DataFrame
        self.df = pd.read_csv(filename, delimiter='\t', header=2)
        self.filename = filename

        # Extract latitude and longitude
        with open(filename, 'r') as file:
            first_line = file.readline().strip()
        data = first_line.split(' ')

        lat_lon = data[6]
        data2 = lat_lon.split(',')
        lat = data2[0]
        lon = data2[1]

        self.lat = lat
        self.lon = lon

        # Extract date
        with open(filename, 'r') as file:
            file.readline()  # Skip the first line
            second_line = file.readline().strip()
        data = second_line.split(' ')
        date = data[1]
        date_string = str(date)
        date_object = datetime.strptime(date_string, '%Y%m%d')
        self.date_only = date_object.date()

        # Define function to convert time range to datetime
        def convert_time_range_to_datetimes(time_range_str, date_str=self.date_only):
            start_time_str, end_time_str = time_range_str.split(' - ')
            start_datetime = datetime.strptime(f'{date_str} {start_time_str}', '%Y-%m-%d %H:%M')
            return pd.Series([start_datetime])
            
        # Apply the function to the DataFrame
        self.df[['date']] = self.df['Periode[UTM]'].apply(lambda x: pd.Series(convert_time_range_to_datetimes(x)))

        self.df.drop(columns=['Periode[UTM]'], inplace=True)

    def concatenate_with(self, other_df, axis=0, ignore_index=False):
        """
        Concatenate the current DataFrame with another DataFrame.

        Parameters:
        - other_df: The other DataFrame to concatenate with.
        - axis: 0 for row-wise concatenation, 1 for column-wise concatenation.
        - ignore_index: Whether to reset the index after concatenation.

        Returns:
        - Concatenated DataFrame.
        """
        return pd.concat([self.df, other_df], axis=axis, ignore_index=ignore_index)
    
    def display_data(self):
        """Display the DataFrame."""
        print(self.df)

# Example usage
filename1 = '/data/IMFSE/PythonCourse/PontdeVilomara/WeatherStations/wx_20220716_CastellouDeBages.csv'
filename2 = '/data/IMFSE/PythonCourse/PontdeVilomara/WeatherStations/wx_20220717_CastellouDeBages.csv'

# Create instances for each file
pdVWs1 = PontdeVilomara(filename1)
pdVWs2 = PontdeVilomara(filename2)

# Concatenate DataFrames from both instances
concatenated_df = pdVWs1.concatenate_with(pdVWs2.df)

# Display the concatenated DataFrame
print(concatenated_df)
```
